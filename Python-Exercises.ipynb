## Decorator

Başka bir fonksiyonu input olarak alırlar ve aldıkları fonksiyonu değiştirmeden ona yeni fonksiyonel özellikler eklerler.

```python
def print_func():
    print('kodluyoruz')

def decorator_func(func):
    def wrapper_func(): 
        print('patika')
        return func()
    
    return wrapper_func
    

    
```

```python
decorated_print = decorator_func(print_func)
decorated_print
```

```python
decorated_print()
```

```python
@decorator_func
def print_func():
    print('patika')
    
print_func()
```

```python
def print_info(name,surname):
    print(f'Name : {name},  Surname : {surname}')
print_info("Ferhat","Metin")
```

```python
def decorater_func(func):
    def wrapper_func(*args):
        print('Kodluyoruz')
        return func(*args)
    return wrapper_func
```

```python
@decorater_func
def print_info(name,surname):
    print(f'Name : {name},  Surname : {surname}')
print_info("Ferhat","Metin")
```

## Class Method ve Static Method

**Class Methodu :** ilk argüman olarak instance yerine class'ı alır ve günceller. 
Instance alsaydı sadece o employee etkilenecekti bu yüzden Employee class'ının içinde class method decorator'ı kullanarak değişikliğin hepsinde olmasını sağlarız.

**Static Method :** Instance veya class'a methodun içerisinde erişim olmuyorsa static olarak tanımlamak daha iyi olabilir. Static methodlar self olarak yani otomatik olarak bir şey vermez.

```python
class Employee:
    raise_percent = 1.05
    num_emp = 0
    def __init__(self,name,last,age,pay):
        self.name = name
        self.last = last
        self.age = age
        self.pay = pay
        Employee.num_emp +=1
        
    def show_info(self):
        print(f'Name : {self.name}, LastName : {self.last}')
        
    def apply_raise(self):
        self.pay = self.pay * Employee.raise_percent
    @classmethod
    def set_raise(self,amount):
        self.raise_percent = amount
        
    @classmethod
    def from_string(cls,emp_str):
        name,last,age,pay = emp_str.split("-")
        return cls(name,last,int(age),int(pay)) # yeni çalışan yaratır ve onu döndürür
    @staticmethod
    def is_dayoff(day):
        if day in ["saturday","sunday"]:
            print("it's a dayoff")
        else:
            print("it is a work day")
        
```

## Magic Method

```python
class Employee:
    raise_percent = 1.05
    num_emp = 0
    def __init__(self,name,last,age,pay):
        self.name = name
        self.last = last
        self.age = age
        self.pay = pay
        Employee.num_emp +=1
        
    def show_info(self):
        print(f'Name : {self.name}, LastName : {self.last}')
        
    def apply_raise(self):
        self.pay = self.pay * self.raise_percent
    def __str__(self):
        return f'Employee(name = {self.name}, = {self.last},age ={self.age},pay = {self.pay})'
    def __add__(self,other):
        return self.pay + other.pay
    def __len__(self):
        return len(self.name)
```

```python
def dict_from_str(s):
    d={}

    for e in s:
        if e not in d:
            d[e]=1
        else:
            d[e]+=1
    return d

def count_balloon(s):
    d_balloon = dict_from_str("BALLOON")
    d_s=dict_from_str(s)
    
    l=[]
    
    for key in d_balloon:
        l.append(d_s[key] // d_balloon[key])
    
    return min(l)
```

```python
def swap_element(a,b):
    element_diff = (sum(a) - sum(b))

    if element_diff % 2 == 0:
        for e_a in a:
            for e_b in b:
                if e_a == e_b + (element_diff / 2):
                    return [e_a, e_b]
                    
                    
    return []
```

```python
a = [1,2,3]
b = [5,2,1]

swap_element(a,b)
```

```python
words=[]
for e in re.split(',* |_|-|!|\+', s):
    if e.isalpha()== True:
        words.append(e)
    if e.isalpha() != True:
        for k in e.split("*"):
            if k.isalpha()==True:
                words.append(k)
            if k.isalpha()!=True:
                for p in k.split("%"):
                    words.append(p)
words_2=[]
for e in words:
    words_2.append(e.lower().capitalize())
"".join(words_2)
    
```
